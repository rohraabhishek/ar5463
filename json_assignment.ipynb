```python
def sep_nested_columns(unst_df,column_name_prefix):
    for i in range(len(unst_df)):
        if type(unst_df[i])==dict:
            index_lst = list(unst_df[i].keys())
            break
        continue
        
    for i in range(len(unst_df)):
        if i!=0:
            if type(unst_df[i])==dict:
                temp_df=pd.DataFrame(unst_df[i],index=[i])
                new_df = pd.concat([new_df,temp_df],axis=0)
            else:
                temp_df=pd.DataFrame(index=[i],columns=index_lst)
                new_df = pd.concat([new_df,temp_df],axis=0)
        else:
            if type(unst_df[0])==dict:
                new_df=pd.DataFrame(unst_df[0],index=[0])
            else:
                new_df=pd.DataFrame(index=[0],columns=index_lst)
                
    final_index_lst = []
    for i in range(len(index_lst)):
        final_index_lst.append(column_name_prefix+index_lst[i])
    new_df.columns = final_index_lst
    
    return new_df
```

```python
import numpy as np
label_lst1=[]
label_lst2=[]
for i in range(len(df2["labels"])):
    if len(df2["labels"][i])==2:
        first,second = (df2['labels'][i])[0],(df2["labels"][i])[1]
    elif len(df2["labels"][i])==1:
        first = (df2['labels'][i])[0]
        second=np.nan
    elif len(df2["labels"][i])==0:
        first=np.nan
        second=np.nan
    label_lst1.append(first)
    label_lst2.append(second)
```

```python
user_df = sep_nested_columns(df2["user"],"user_")
```

```python
label_lst1_df = sep_nested_columns(label_lst1,"label_lst1_")
```

```python
label_lst2_df = sep_nested_columns(label_lst2,"label_lst2_")
```

```python
pull_request_df = sep_nested_columns(df2["pull_request"],"pull_request_")
```
